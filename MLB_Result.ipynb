```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
sportUrl = 'https://s5.sir.sportradar.com/taiwansportslottery/zht/3/season/88607'
response = requests.get(sportUrl)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table = soup.select_one('.panel-body .visible-md-up .table')
```

```python
headThs = table.select('thead tr th')
```

```python
list1 = []

for th in headThs:
    list1.append(th.text)
```

```python
heads = [th.text for th in headThs]
heads
```

```python
bodyTr = table.select('tbody tr')
len(bodyTr)
```

```python
tr = bodyTr[0]
tds = [td.text for td in tr.select('td')]
tds
```

```python
tBodyData  = []
for tr in bodyTr:
    trData = {}
    for index, td in enumerate(tr.select('td')):
        text = td.text
        if index == 1:
            text = text[:len(text)//2]
        trData[heads[index]] = text
    tBodyData.append(trData)
tBodyData
```

```python
df = pd.DataFrame(tBodyData)
df
```

```python
df.to_csv('./table.csv', index=False)
```
