```python
import requests
from bs4 import BeautifulSoup
import pandas

def get_FinDetail(stk_id):
    headers = {
        'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Mobile Safari/537.36'
    }
    res = requests.get('https://goodinfo.tw/tw/StockBzPerformance.asp?STOCK_ID='+stk_id, headers = headers)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text, 'lxml')
    data = soup.select_one('#txtFinDetailData')
    dfs = pandas.read_html(data.prettify())
    dfs[0].columns = dfs[0].columns.get_level_values(1)
    return dfs[0]
```

```python
stock = input('Please input a stock id: ')

get_FinDetail(stock)
```
